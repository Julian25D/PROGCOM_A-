```python
import matplotlib.pyplot as plt
from IPython.display import display, HTML, clear_output
import time

# Función para determinar el rango de saturación de oxígeno
def determinar_rango_spo2(spo2):
    if spo2 >= 95:
        return "Normal (95% - 100%)", 'green'
    elif 90 <= spo2 < 95:
        return "Levemente baja (90% - 94%)", 'orange'
    else:
        return "Baja (< 90%)", 'red'

# Función para mostrar la interfaz del monitor
def mostrar_monitor(spo2, tomas):
    # Determinar el rango y el color
    rango, color = determinar_rango_spo2(spo2)

    # Limpiar la salida anterior para simular actualización
    clear_output(wait=True)

    # Crear la visualización tipo "monitor"
    plt.figure(figsize=(6, 2))
    plt.barh(["Saturación de Oxígeno"], [spo2], color=color, alpha=0.7)

    # Mostrar texto con el valor y el rango
    plt.text(spo2 + 2, 0, f"SpO2: {spo2}%", fontsize=12, va='center', color='white', fontweight='bold')
    plt.text(spo2 + 2, -0.2, f"Rango: {rango}", fontsize=12, va='center', color='white', fontweight='bold')

    # Etiquetas y título
    plt.title("Monitor de Saturación de Oxígeno", fontsize=14)
    plt.xlim(0, 100)
    plt.yticks([])  # Eliminar las marcas en el eje Y
    plt.xlabel("Porcentaje de SpO2", fontsize=12)

    # Si la saturación es baja (< 90%), emitir una alerta visual
    if color == 'red':
        display(HTML("<script>alert('¡ALERTA! La saturación de oxígeno es baja (< 90%)')</script>"))

    # Mostrar el gráfico
    plt.show()

    # Mostrar el historial hasta el momento
    print("\nHistorial de tomas de SpO2:")
    for idx, toma in enumerate(tomas, 1):
        print(f"Toma {idx}: {toma}%")

# Lista para almacenar las tomas de SpO2
tomas_spo2 = []

# Simulación de la actualización del monitor
while True:
    try:
        # Pedir al usuario que ingrese la saturación de oxígeno
        spo2 = float(input("Ingrese la saturación de oxígeno (SpO2) en porcentaje: "))

        # Agregar la toma a la lista
        tomas_spo2.append(spo2)

        # Mostrar el monitor con la saturación ingresada
        mostrar_monitor(spo2, tomas_spo2)

        # Si tenemos 10 tomas, mostrar el historial completo y finalizar
        if len(tomas_spo2) >= 10:
            print("\n¡Se han alcanzado 10 tomas! Mostrando el historial completo:")
            print("\nHistorial de 10 últimas tomas de SpO2:")
            for idx, toma in enumerate(tomas_spo2, 1):
                print(f"Toma {idx}: {toma}%")
            break  # Finalizamos el ciclo después de mostrar el historial

        # Pausar por 2 segundos antes de permitir otro ingreso
        time.sleep(2)

    except ValueError:
        print("Por favor, ingrese un número válido.")

```

```python
import matplotlib.pyplot as plt
import numpy as np
import datetime
import pytz

# Función para determinar el rango de saturación de oxígeno
def determinar_rango_spo2(spo2):
    if spo2 >= 95:
        return 'green'  # Normal
    elif 90 <= spo2 < 95:
        return 'orange'  # Levemente bajo
    elif 80 <= spo2 < 90:
        return 'red'  # Bajo
    else:
        return 'purple'  # Crítico, SpO2 < 80

# Lista para almacenar las tomas de SpO2 y las horas
tomas_spo2 = []
horas_tomas = []

# Establecer la zona horaria de Colombia (Bogotá)
colombia_tz = pytz.timezone('America/Bogota')

# Simulación de la actualización del monitor con mediciones infinitas
while True:
    try:
        # Pedir al usuario que ingrese la saturación de oxígeno
        entrada = input("Ingrese la saturación de oxígeno (SpO2) en porcentaje o escriba 'salir' para terminar: ")

        if entrada.lower() == 'salir':
            break  # Romper el bucle y finalizar el proceso

        # Convertir el valor ingresado en número flotante
        spo2 = float(entrada)

        # Obtener la hora actual en la zona horaria de Colombia
        hora_actual = datetime.datetime.now(colombia_tz).strftime("%H:%M:%S")

        # Agregar la toma y la hora a las listas
        tomas_spo2.append(spo2)
        horas_tomas.append(hora_actual)

        # Mostrar el valor y el rango de la última toma
        rango_color = determinar_rango_spo2(spo2)
        print(f"Toma {len(tomas_spo2)} a las {hora_actual}: {spo2}% - Color: {rango_color}")

    except ValueError:
        print("Por favor, ingrese un número válido o 'salir' para terminar.")

# Al final, mostrar el historial completo de mediciones
print("\n¡El proceso ha terminado! Mostrando el historial completo:")
for idx, (hora, toma) in enumerate(zip(horas_tomas, tomas_spo2), 1):
    print(f"Toma {idx} a las {hora}: {toma}%")

# Convertir la lista de SpO2 a un array de numpy para hacer más fácil la manipulación
tomas_spo2 = np.array(tomas_spo2)

# Crear una figura para el gráfico
plt.figure(figsize=(10, 6))

# Graficar la línea negra que une los puntos
plt.plot(tomas_spo2, color='black', lw=2, zorder=1)

# Graficar los puntos individuales con colores según el rango
for i in range(len(tomas_spo2)):
    color = determinar_rango_spo2(tomas_spo2[i])
    plt.scatter(i, tomas_spo2[i], color=color, s=80, zorder=5)  # Puntos grandes

# Agregar títulos y etiquetas al gráfico
plt.title("Monitoreo de Saturación de Oxígeno de usuario ", fontsize=14)
plt.xlabel("Hora de la medición", fontsize=12)
plt.ylabel("Porcentaje de SpO2", fontsize=12)

# Reemplazar el índice por las horas reales en el eje X
plt.xticks(ticks=np.arange(len(horas_tomas)), labels=horas_tomas, rotation=45, ha='right')

# Ajustar los límites del eje X y Y para dejar espacio
plt.xlim(-0.5, len(tomas_spo2) - 0.5)  # Espacio en ambos extremos del eje X
plt.ylim(70, 100)  # Espacio adicional en el eje Y para ver los puntos sin que toquen el borde

# Mostrar leyenda con los colores
plt.scatter([], [], color='green', label='Normal (95%-100%)')
plt.scatter([], [], color='orange', label='Levemente baja (90%-94%)')
plt.scatter([], [], color='red', label='Baja (80%-89%)')
plt.scatter([], [], color='purple', label='Crítico (<80%)')
plt.legend()

# Mostrar el gráfico final
plt.grid(True)
plt.tight_layout()  # Ajusta la disposición para que las etiquetas del eje X no se sobrepongan
plt.show()

```
