```python
# Diccionarios de emojis
emojis_masa = {
    "delgada": "🌮",
    "gruesa": "🍞",
    "sin gluten": "🥦"
}

emojis_salsa = {
    "tomate": "🍅",
    "bbq": "🥩",
    "blanca": "🥛"
}

emojis_queso = {
    "mozzarella": "🧀",
    "vegano": "🌱",
    "extra queso": "🧀🧀"
}

emojis_coccion = {
    "normal": "🔥",
    "extra cocida": "♨️",
    "poco cocida": "💧"
}

emojis_forma = {
    "redonda": "⚪",
    "cuadrada": "🟥",
    "estrella": "⭐"
}

emojis_ingredientes = {
    "pepperoni": "🍕",
    "jamón": "🥓",
    "queso": "🧀",
    "champiñones": "🍄",
    "piña": "🍍",
    "aceitunas": "🫒",
    "pimiento": "🌶️",
    "cebolla": "🧅"
}

# Mostrar carta
def mostrar_carta():
    print("\n📜 BIENVENIDO A LA PIZZERÍA CREATIVA 🍕")
    print("Aquí está nuestra carta. ¡Escoge lo que más te guste!\n")

    print("🥖 MASAS DISPONIBLES:")
    for masa, emoji in emojis_masa.items():
        print(f"  - {masa.title()} {emoji}")

    print("\n🍅 SALSAS DISPONIBLES:")
    for salsa, emoji in emojis_salsa.items():
        print(f"  - {salsa.title()} {emoji}")

    print("\n🧀 QUESOS DISPONIBLES:")
    for queso, emoji in emojis_queso.items():
        print(f"  - {queso.title()} {emoji}")

    print("\n🍄 INGREDIENTES ADICIONALES:")
    for ing, emoji in emojis_ingredientes.items():
        print(f"  - {ing.title()} {emoji}")

    print("\n🔥 TIPOS DE COCCIÓN:")
    for coccion, emoji in emojis_coccion.items():
        print(f"  - {coccion.title()} {emoji}")

    print("\n🟠 FORMAS DE LA PIZZA:")
    for forma, emoji in emojis_forma.items():
        print(f"  - {forma.title()} {emoji}")
    print("\n" + "-"*40 + "\n")

# Clase Pedido
class Pedido:
    def __init__(self, masa, salsa, queso, ingredientes, coccion, forma):
        self.masa = masa
        self.salsa = salsa
        self.queso = queso
        self.ingredientes = ingredientes
        self.coccion = coccion
        self.forma = forma

    def mostrar(self):
        print("\n🧾 Pedido del cliente:")
        print(f"Masa: {self.masa} {emojis_masa.get(self.masa, '❓')}")
        print(f"Salsa: {self.salsa} {emojis_salsa.get(self.salsa, '❓')}")
        print(f"Queso: {self.queso} {emojis_queso.get(self.queso, '❓')}")
        ing_emojis = [emojis_ingredientes.get(i, '❓') for i in self.ingredientes]
        print(f"Ingredientes: {', '.join(self.ingredientes)} {' '.join(ing_emojis)}")
        print(f"Cocción: {self.coccion} {emojis_coccion.get(self.coccion, '❓')}")
        print(f"Forma: {self.forma} {emojis_forma.get(self.forma, '❓')}")

# Función para ingresar pedido
def ingresar_pedido():
    masa = input("🍞 Tipo de masa: ").lower()
    salsa = input("🍅 Tipo de salsa: ").lower()
    queso = input("🧀 Tipo de queso: ").lower()

    print("🧂 Ingredientes (separados por comas):")
    ingredientes = [i.strip().lower() for i in input().split(",")]

    coccion = input("🔥 Tipo de cocción: ").lower()
    forma = input("🟠 Forma de la pizza: ").lower()

    return Pedido(masa, salsa, queso, ingredientes, coccion, forma)

# Juego principal
def jugar_pizzeria():
    print("🍕 ¡Bienvenido a tu Pizzería Creativa! 🍕")
    mostrar_carta()
    n = int(input("👥 ¿Cuántos comensales hay hoy?: "))
    pedidos = []

    for i in range(n):
        print(f"\n👨‍🍳 --- Pedido del Cliente {i+1} ---")
        pedido = ingresar_pedido()
        pedido.mostrar()
        pedidos.append(pedido)

    print("\n✅ ¡Todos los pedidos han sido registrados! 🧑‍🍳 Ahora puedes prepararlos uno a uno.")

# Ejecutar juego
if __name__ == "__main__":
    jugar_pizzeria()
```
