```python
# Diccionarios de emojis
emojis_masa = {
    "delgada": "ğŸŒ®",
    "gruesa": "ğŸ",
    "sin gluten": "ğŸ¥¦"
}

emojis_salsa = {
    "tomate": "ğŸ…",
    "bbq": "ğŸ¥©",
    "blanca": "ğŸ¥›"
}

emojis_queso = {
    "mozzarella": "ğŸ§€",
    "vegano": "ğŸŒ±",
    "extra queso": "ğŸ§€ğŸ§€"
}

emojis_coccion = {
    "normal": "ğŸ”¥",
    "extra cocida": "â™¨ï¸",
    "poco cocida": "ğŸ’§"
}

emojis_forma = {
    "redonda": "âšª",
    "cuadrada": "ğŸŸ¥",
    "estrella": "â­"
}

emojis_ingredientes = {
    "pepperoni": "ğŸ•",
    "jamÃ³n": "ğŸ¥“",
    "queso": "ğŸ§€",
    "champiÃ±ones": "ğŸ„",
    "piÃ±a": "ğŸ",
    "aceitunas": "ğŸ«’",
    "pimiento": "ğŸŒ¶ï¸",
    "cebolla": "ğŸ§…"
}

# Mostrar carta
def mostrar_carta():
    print("\nğŸ“œ BIENVENIDO A LA PIZZERÃA CREATIVA ğŸ•")
    print("AquÃ­ estÃ¡ nuestra carta. Â¡Escoge lo que mÃ¡s te guste!\n")

    print("ğŸ¥– MASAS DISPONIBLES:")
    for masa, emoji in emojis_masa.items():
        print(f"  - {masa.title()} {emoji}")

    print("\nğŸ… SALSAS DISPONIBLES:")
    for salsa, emoji in emojis_salsa.items():
        print(f"  - {salsa.title()} {emoji}")

    print("\nğŸ§€ QUESOS DISPONIBLES:")
    for queso, emoji in emojis_queso.items():
        print(f"  - {queso.title()} {emoji}")

    print("\nğŸ„ INGREDIENTES ADICIONALES:")
    for ing, emoji in emojis_ingredientes.items():
        print(f"  - {ing.title()} {emoji}")

    print("\nğŸ”¥ TIPOS DE COCCIÃ“N:")
    for coccion, emoji in emojis_coccion.items():
        print(f"  - {coccion.title()} {emoji}")

    print("\nğŸŸ  FORMAS DE LA PIZZA:")
    for forma, emoji in emojis_forma.items():
        print(f"  - {forma.title()} {emoji}")
    print("\n" + "-"*40 + "\n")

# Clase Pedido
class Pedido:
    def __init__(self, masa, salsa, queso, ingredientes, coccion, forma):
        self.masa = masa
        self.salsa = salsa
        self.queso = queso
        self.ingredientes = ingredientes
        self.coccion = coccion
        self.forma = forma

    def mostrar(self):
        print("\nğŸ§¾ Pedido del cliente:")
        print(f"Masa: {self.masa} {emojis_masa.get(self.masa, 'â“')}")
        print(f"Salsa: {self.salsa} {emojis_salsa.get(self.salsa, 'â“')}")
        print(f"Queso: {self.queso} {emojis_queso.get(self.queso, 'â“')}")
        ing_emojis = [emojis_ingredientes.get(i, 'â“') for i in self.ingredientes]
        print(f"Ingredientes: {', '.join(self.ingredientes)} {' '.join(ing_emojis)}")
        print(f"CocciÃ³n: {self.coccion} {emojis_coccion.get(self.coccion, 'â“')}")
        print(f"Forma: {self.forma} {emojis_forma.get(self.forma, 'â“')}")

# FunciÃ³n para ingresar pedido
def ingresar_pedido():
    masa = input("ğŸ Tipo de masa: ").lower()
    salsa = input("ğŸ… Tipo de salsa: ").lower()
    queso = input("ğŸ§€ Tipo de queso: ").lower()

    print("ğŸ§‚ Ingredientes (separados por comas):")
    ingredientes = [i.strip().lower() for i in input().split(",")]

    coccion = input("ğŸ”¥ Tipo de cocciÃ³n: ").lower()
    forma = input("ğŸŸ  Forma de la pizza: ").lower()

    return Pedido(masa, salsa, queso, ingredientes, coccion, forma)

# Juego principal
def jugar_pizzeria():
    print("ğŸ• Â¡Bienvenido a tu PizzerÃ­a Creativa! ğŸ•")
    mostrar_carta()
    n = int(input("ğŸ‘¥ Â¿CuÃ¡ntos comensales hay hoy?: "))
    pedidos = []

    for i in range(n):
        print(f"\nğŸ‘¨â€ğŸ³ --- Pedido del Cliente {i+1} ---")
        pedido = ingresar_pedido()
        pedido.mostrar()
        pedidos.append(pedido)

    print("\nâœ… Â¡Todos los pedidos han sido registrados! ğŸ§‘â€ğŸ³ Ahora puedes prepararlos uno a uno.")

# Ejecutar juego
if __name__ == "__main__":
    jugar_pizzeria()
```
