```python
# Clase que simula una cuenta bancaria con encapsulamiento
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        # El saldo estÃ¡ encapsulado, solo puede ser accedido o modificado a travÃ©s de mÃ©todos
        self.__saldo = saldo_inicial

    # MÃ©todo getter para obtener el saldo
    def get_saldo(self):
        return self.__saldo

    # MÃ©todo setter para modificar el saldo (depÃ³sito)
    def depositar(self, cantidad):
        if cantidad > 0:
            self.__saldo += cantidad
            print(f"DepÃ³sito exitoso: ${cantidad}")
        else:
            print("Cantidad de depÃ³sito no vÃ¡lida.")

    # MÃ©todo setter para retirar dinero
    def retirar(self, cantidad):
        if cantidad > 0 and cantidad <= self.__saldo:
            self.__saldo -= cantidad
            print(f"Retiro exitoso: ${cantidad}")
        else:
            print("Saldo insuficiente o cantidad invÃ¡lida.")

    # MÃ©todo para mostrar informaciÃ³n de la cuenta
    def mostrar_info(self):
        print(f"\n--- InformaciÃ³n de la cuenta ---")
        print(f"Titular: {self.titular}")
        print(f"Saldo actual: ${self.__saldo}")
        print("----------------------------------")

# FunciÃ³n para interactuar dinÃ¡micamente con el usuario
def interactuar_con_cuenta():
    # Datos iniciales
    titular = input("Introduce el nombre del titular de la cuenta: ")
    saldo_inicial = float(input("Introduce el saldo inicial de la cuenta: $"))

    # Creamos una instancia de CuentaBancaria
    cuenta = CuentaBancaria(titular, saldo_inicial)

    while True:
        # Mostramos el menÃº de opciones
        print("\nÂ¿QuÃ© te gustarÃ­a hacer?")
        print("1. Consultar saldo")
        print("2. Depositar dinero")
        print("3. Retirar dinero")
        print("4. Ver informaciÃ³n de la cuenta")
        print("5. Salir")

        # Obtenemos la opciÃ³n del usuario
        opcion = input("Selecciona una opciÃ³n (1-5): ")

        if opcion == '1':
            print(f"Saldo actual: ${cuenta.get_saldo()}")

        elif opcion == '2':
            cantidad = float(input("Introduce la cantidad a depositar: $"))
            cuenta.depositar(cantidad)

        elif opcion == '3':
            cantidad = float(input("Introduce la cantidad a retirar: $"))
            cuenta.retirar(cantidad)

        elif opcion == '4':
            cuenta.mostrar_info()

        elif opcion == '5':
            print("Â¡Gracias por usar el sistema de cuentas bancarias!")
            break

        else:
            print("OpciÃ³n no vÃ¡lida. Por favor selecciona una opciÃ³n entre 1 y 5.")

# Ejecutamos la interacciÃ³n dinÃ¡mica
interactuar_con_cuenta()
```
